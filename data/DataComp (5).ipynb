```python
def noComma(num):
    ans = ""
    for i in num:
        if i != ",":
            ans+= i
    return int(ans)

def percentChange(val1, val2):
    val2 = noComma(val2)
    val1 = noComma(val1)
    return (int(val2)/int(val1) - 1)*100

print(percentChange(df["DNK"][110], df["DNK"][129]))
    
```
